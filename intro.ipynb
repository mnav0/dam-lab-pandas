# Finding and updating

```python
# locate a row, delete it
    #are there any rows with notes that we want to delete?
squirrels_df[squirrels_df['Color notes'] != ""]
```

```python
    # maybe get rid of "just outside hectare" row - this squirrel may not be accurately recorded.
    #location is 6
squirrels_df.iloc[6]
```

```python
#and drop it
squirrels_df.drop(index=6, inplace=True)
```

```python
#check that it's gone
squirrels_df.iloc[4:8]
```

```python
# condense all of the boolean values into one "Action" column

#possible actions:
actions = ['Running', 'Chasing', 'Climbing', 'Eating', 'Foraging']
#if a squirrel has a "True" in action category, add that category to the "Action" column. If a squirrel has multiple "True" values, concatenate those categories together in the "Action" column.
def concatenate_actions(row):
    action_list = []
    for action in actions:
        if row[action] == True:
            action_list.append(action)
    return ", ".join(action_list)
squirrels_df['Action'] = squirrels_df.apply(concatenate_actions, axis=1)


squirrels_df.head()
```

# Grouping

```python
# aggregate by a column

#how many squirrels of each primary fur color are there, in AM vs PM?
squirrels_df.groupby(['Primary Fur Color', 'Shift']).size()
```
